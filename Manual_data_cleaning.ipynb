```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/PNRao_SalesRetail_Sales_Transactions.csv', on_bad_lines='skip')
df.head()
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = df['Date'].fillna(df['Date'].mode()[0])
```

```python
df['PromotionID'] = df['PromotionID'].fillna('PROMO-00')
```

```python
df['TotalAmount'] = pd.to_numeric(df['TotalAmount'], errors='coerce')
df['TotalAmount'] = df['TotalAmount'].fillna(df['TotalAmount'].mean())
```

```python
df['TransactionID'] = df['TransactionID'].str.strip()
print(df['TransactionID'])
```

```python
df['CustomerID'] = df['CustomerID'].str.strip()
print(df['CustomerID'])
```

```python
df['StoreID'] = df['StoreID'].str.strip()
print(df['StoreID'])
```

```python
df['EmployeeID'] = df['EmployeeID'].str.strip()
print(df['EmployeeID'])
```

```python
df['PromotionID'] = df['PromotionID'].str.strip()
print(df['PromotionID'])
```

```python
df['PaymentMethod'] = df['PaymentMethod'].str.strip()
print(df['PaymentMethod'])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'])
```

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
print("\nShape:")
print(df.shape)
```
